```python
import xlwings as xw
# input path excel
filepath="Daily-update-09.08.2023.xlsx"
#getconnection
def getbook(filepath):
    booktoexcel=xw.Book(filepath)
    return booktoexcel
#get sheet
def getsheet(excel_book):
    sheet=excel_book.sheets("Sheet1")
    return sheet
#get end of row and column
def find_last_row(data_sheet):
    end_row=data_sheet.range('A2').end('down').row
    return end_row
def find_last_column(data_sheet):
    end_column=data_sheet.range('A2').end('right').column
    return end_column
#getdata
def getdata(r0,c0,r1,c1,sheet):
    data=sheet.range((r0,c0),(r1,c1)).value
    return data
def printdata(r,c,sheet,data):
     sheet.range((r,c)).value=data

# call funtion
daily_excel=getbook(filepath)
sht1_daily=getsheet(daily_excel)
daily_r=find_last_row(sht1_daily)
daily_c=find_last_column(sht1_daily)

for x in range(1,daily_r,1):
    i=x+1
    
      
    try:
         #get name file to update
        get_name_file=getdata(i,1,i,1,sht1_daily)
        #getconnection to excel
        data_update=getbook(get_name_file+".xlsx")
    except:
        print("không tìm thấy file để update")
        print("kiểm tra lại file ")
        break    
    #get sheet
    data_sheet=getsheet(data_update)
    #get end of row and column to file update
    end_row=find_last_row(data_sheet)
    end_column=find_last_column(data_sheet)
    datacompare=getdata(i,3,i,3,sht1_daily)
    
    if datacompare != getdata(3,3,3,3,data_sheet) : #and datacompare > getdata(3,3,3,3,data_sheet)    
        #get 1 row of daily
        data=getdata(i,1,i,daily_c,sht1_daily)
        #get data of file need update
        data_update_getdata=getdata(3,1,end_row,end_column,data_sheet)
        # update data new 
        printdata(3,1,data_sheet,data)
        # update data old
        printdata(4,1,data_sheet,data_update_getdata)
        datasheet_r=data_sheet.range("A"+str(data_sheet.cells.last_cell.row)).end("up").row
        data_sheet.range((datasheet_r,1),(datasheet_r,end_column)).delete()
        # turn on to save file auto
        # data_update.save()
        # data_update.close()
    else: 
        print("trùng dữ liệu tại file "+get_name_file)
        print("nhập 0 để dừng")
        a=input()
        if(a==0):
            break
     
 
        
    
                

    
        
    
    
    
    
    
    
   
   
   
    
   
    
    
    



```
